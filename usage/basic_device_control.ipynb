# Basic device control

This notebook shows the core control flow for a single Venta protocol v2 device. The code was tested agains an LW74 device.

```python
# this is for testing (as long as the package is not loadable using pip)
import sys
sys.path.insert(0, "../")    # relative to notebook dir
import venta_protocol_v2_device
```

```python
from venta_protocol_v2_device import Venta_Protocol_v2_Device

DEVICE_IP = "10.0.XX.YY"  # Replace with your device IP
device = Venta_Protocol_v2_Device(DEVICE_IP)
device   # show the object
```

```python
status = device.getStatus()
status
```

## Example controls

Example Python controling the Venta device.

```python
device.setPower(True)
```

```python
device.setAutomatic(True)
```

```python
device.setFanSpeed(1)
```

```python
device.setTargetHum(50)
```

```python
device.setLEDStripColor("#080808")
```

```python
device.setLEDStripActive(True)
```

```python
device.setSleepMode(True)
```

```python
device.setLEDStripMode("3")
```

Note: if an argument type is not proper, the call will raise a meaningfull error message:

```python
device.setFanSpeed("some incorrect type")
```
